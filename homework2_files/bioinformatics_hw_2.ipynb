###1

Перевели транскрипты в белковые последовательности с помощью TransDecoder.LongOrfs

```
TransDecoder.LongOrfs -t transcripts.fasta
```



###2

 Создали базу данных BLAST на основе транскриптов, переведенных в белковые последовательности с помощью makeblastdb

```
makeblastdb \
  -in transcriptoms.fasta.transdecoder_dir/longest_orfs.pep \
  -dbtype prot \
  -out luciferase_candidates_db \
  -parse_seqids
```



###3

Нашли вхождения люцеферазы с помощью BLAST


```
blastp \
  -query luciola_luciferase.faa \
  -db luciferase_candidates_db \
  -out luciferase_vs_candidates.blastp \
  -evalue 1e-5 \
  -max_target_seqs 50 \
  -outfmt "6 qseqid sseqid pident length mismatch gapopen qstart qend sstart send evalue bitscore qlen slen"
```
Для поиска гомологов люциферазы в наборе предсказанных белков был использован инструмент blastp. В качестве query использовалась аминокислотная последовательность люциферазы Luciola cruciata, а в качестве database — белковая база, созданная из ORF, предсказанных TransDecoder. Хиты с e‑value выше 1e‑5 отбрасывались. Это позволило исключить слабые и нерелевантные совпадения. max_target_seqs 50 — ограничивал максимальное число возвращаемых совпадений.




###4

Выбрали 10 лучших хитов:


```
sort -k11,11g -k12,12gr luciferase_vs_candidates.blastp > luciferase_vs_candidates.sorted.blastp
```
Сортируем по 11 колонке - значения e-value - по возрастанию, в случае если e-value одинаковые сортируем по12 колонке - bitscore - по убыванию


```
head -n 10 luciferase_vs_candidates.sorted.blastp > luciferase_top10_hits.blastp
```
Отбираем первые 10
